```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('NSE MNC REVENUE 21-23.csv')
```

```python
display(data)
```

```python
data.dropna(axis=1, how='all', inplace=True)
data.dropna(axis=0, how='all', inplace=True)
display(data)
```

```python
print("Missing Values:")
print(data.isnull().sum())
```

```python
data.columns = ['S.No.', 'Company Name', 'Company Email', 'FY23', 'FY22', 'FY21']

display(data)
```

```python
data['Company Name'] = data['Company Name'].astype(str)
```

```python
revenue_cols = ['FY21', 'FY22', 'FY23']

non_numeric_values = {}
for col in revenue_cols:
    non_numeric_rows = data[pd.to_numeric(data[col], errors='coerce').isnull()]
    if not non_numeric_rows.empty:
        non_numeric_values[col] = non_numeric_rows[col].unique()
```

```python
if non_numeric_values:
    print("\nNon-numeric values in revenue columns:")
    for col, values in non_numeric_values.items():
        print(f"Column '{col}': {values}")
```

```python
data['FY21_FY22_Change'] = data['FY22'] - data['FY21']
data['FY22_FY23_Change'] = data['FY23'] - data['FY22']
```

```python
data[revenue_cols] = data[revenue_cols].apply(pd.to_numeric, errors='coerce')
```

```python
data['FY21_FY22_Change'] = data['FY22'] - data['FY21']
data['FY22_FY23_Change'] = data['FY23'] - data['FY22']
```

```python
print("\nYear-to-Year Changes:")
display(data[['Company Name', 'FY21_FY22_Change', 'FY22_FY23_Change']])
```

```python
data['Profitable'] = (data['FY23'] > data['FY21'])

plt.figure(figsize=(8, 6))
plt.scatter(data['FY21_FY22_Change'], data['FY22_FY23_Change'])
plt.xlabel('FY21 to FY22 Change')
plt.ylabel('FY22 to FY23 Change')
plt.title('Year-to-Year Comparison: FY21 to FY22 vs. FY22 to FY23')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['FY22_FY23_Change'], data['FY21_FY22_Change'])
plt.xlabel('FY22 to FY23 Change')
plt.ylabel('FY21 to FY22 Change')
plt.title('Year-to-Year Comparison: FY22 to FY23 vs. FY21 to FY22')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(data.index, data['FY21'], color='purple', alpha=0.5, label='FY21')
plt.bar(data.index, data['FY22'], color='yellow', alpha=0.5, label='FY22')
plt.xticks(range(len(data['Company Name'])), data['Company Name'], rotation=90)
plt.xlabel('Company Name')
plt.ylabel('Revenue')
plt.title('Comparison of Revenues in FY21 and FY22')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(data.index, data['FY22'], color='purple', alpha=0.5, label='FY22')
plt.bar(data.index, data['FY23'], color='yellow', alpha=0.5, label='FY23')
plt.xticks(range(len(data['Company Name'])), data['Company Name'], rotation=90)
plt.xlabel('Company Name')
plt.ylabel('Revenue')
plt.title('Comparison of Revenues in FY22 and FY23')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(data.index, data['FY21'], color='blue', alpha=0.5, label='FY21')
plt.bar(data.index, data['FY23'], color='red', alpha=0.5, label='FY23')
plt.xticks(range(len(data['Company Name'])), data['Company Name'], rotation=90)
plt.xlabel('Company Name')
plt.ylabel('Revenue')
plt.title('Comparison of Revenues in FY21 and FY23')
plt.legend()
plt.show()
```

```python
# 1. Positive Changes
positive_changes = data[(data['FY21_FY22_Change'] > 0) & (data['FY22_FY23_Change'] > 0)]

# 2. Negative Changes
negative_changes = data[(data['FY21_FY22_Change'] < 0) | (data['FY22_FY23_Change'] < 0)]

# 3. Missing Values
missing_values = data[data.isnull().any(axis=1)]

print("Positive Changes:")
display(positive_changes)
print("\nNegative Changes:")
display(negative_changes)
print("\nMissing Values:")
display(missing_values)
```

```python
data['% Change FY21-FY22'] = ((data['FY22'] - data['FY21']) / data['FY21']) * 100
data['% Change FY22-FY23'] = ((data['FY23'] - data['FY22']) / data['FY22']) * 100
print("Data with Percentage Change in Revenue:")
display(data[['Company Name','FY21_FY22_Change', '% Change FY21-FY22', 'FY22_FY23_Change', '% Change FY22-FY23']])
```

```python
plt.figure(figsize=(10, 6))

plt.bar(data['Company Name'], data['% Change FY21-FY22'], color='blue', alpha=0.5, label='% Change FY21-FY22')
plt.bar(data['Company Name'], data['% Change FY22-FY23'], color='red', alpha=0.5, label='% Change FY22-FY23')

plt.xlabel('Company')
plt.ylabel('% Change in Revenue')
plt.title('Percentage Change in Revenue for Each Company')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()

plt.show()
```

```python
print("\nCompanies with Best Change in Revenue Amount (FY21-FY22):")
best_change_21_22 = data.nlargest(5, 'FY21_FY22_Change')
display(best_change_21_22[['Company Name', 'FY21_FY22_Change']])

print("\nCompanies with Worst Change in Revenue Amount (FY21-FY22):")
worst_change_21_22 = data.nsmallest(5, 'FY21_FY22_Change')
display(worst_change_21_22[['Company Name', 'FY21_FY22_Change']])

print("\nCompanies with Best Change in Revenue Amount (FY22-FY23):")
best_change_22_23 = data.nlargest(5, 'FY22_FY23_Change')
display(best_change_22_23[['Company Name', 'FY22_FY23_Change']])

print("\nCompanies with Worst Change in Revenue Amount (FY22-FY23):")
worst_change_22_23 = data.nsmallest(5, 'FY22_FY23_Change')
display(worst_change_22_23[['Company Name', 'FY22_FY23_Change']])
```

```python
print("\nCompanies with Best Change in Revenue % (FY21-FY22):")
best_change_21_22 = data.nlargest(5, '% Change FY21-FY22')
display(best_change_21_22[['Company Name', '% Change FY21-FY22']])

print("\nCompanies with Worst Change in Revenue % (FY21-FY22):")
worst_change_21_22 = data.nsmallest(5, '% Change FY21-FY22')
display(worst_change_21_22[['Company Name', '% Change FY21-FY22']])

print("\nCompanies with Best Change in Revenue % (FY22-FY23):")
best_change_22_23 = data.nlargest(5, '% Change FY22-FY23')
display(best_change_22_23[['Company Name', '% Change FY22-FY23']])

print("\nCompanies with Worst Change in Revenue % (FY22-FY23):")
worst_change_22_23 = data.nsmallest(5, '% Change FY22-FY23')
display(worst_change_22_23[['Company Name', '% Change FY22-FY23']])
```

```python
plt.figure(figsize=(10, 6))

for j, row in data.iterrows():
    height = row['% Change FY21-FY22']
    plt.bar(j, height, color='blue', alpha=0.5)
    plt.text(j, 0, f"{height:.1f}%", ha='center', va='bottom', rotation=90)

plt.xticks(range(len(data['Company Name'])), data['Company Name'], rotation=90)
plt.xlabel('Company')
plt.ylabel('% Change in Revenue')
plt.title('Percentage Change in Revenue (FY21-FY22)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

for j, row in data.iterrows():
    height = row['% Change FY22-FY23']
    plt.bar(j, height, color='red', alpha=0.5)
    plt.text(j, 0, f"{height:.2f}%", ha='center', va='bottom', rotation=90)

plt.xticks(range(len(data['Company Name'])), data['Company Name'], rotation=90)
plt.xlabel('Company')
plt.ylabel('% Change in Revenue')
plt.title('Proportional Area Graph for Percentage Change in Revenue (FY22-FY23)')
plt.tight_layout()
plt.show()
```
